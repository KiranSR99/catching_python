### Filtering

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
df['sepal_length'] <= 5.0
```

```python
fdf = df[df['sepal_length'] <= 5.0]
```

```python
fdf
```

```python
fdf.info()
```

```python
# filtering with multiple conditions
(df['sepal_length'] <=5) & (df['petal_length'] <=3.0)
```

```python
fdf2 = df[(df['sepal_length'] <=5) & (df['petal_length'] <=3.0)]
```

```python
fdf3 = df[(df['sepal_width'] == 3.0 ) & (df['petal_width'] == 0.2)]
```

```python
fdf3
```

```python
# filtering by string matching
df[df['class'].str.contains('setosas')]
```

```python
df['class'].unique()
```

```python
# filer by is in
l1 = ['Iris-setosa', 'Iris-versicolor']
df[df['class'].isin(l1)]
```

```python
# filtering by null values => isnull(), isna(), notnull(), notna()

df[df.isna()]
df[df.notnull()]
```

```python
# query
df.query('sepal_width == 3.0 and petal_width == 0.2')
```
