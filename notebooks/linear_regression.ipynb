### Selecting Target and Fetures

```python
# our target variable is price

y = df.Price
```

```python
df.columns
```

```python
features = ['Rooms', 'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'YearBuilt']
```

```python
# features variable
X = df[features]
```

```python
X.corr()
```

### Build a model

```python
# importing scikit-learn
from sklearn.tree import DecisionTreeRegressor
```

```python
# building the model
house_model = DecisionTreeRegressor(random_state=1)
```

```python
# fit the data into the model
house_model.fit(X, y)
```

```python
# making predictions
X.head()
```

```python
house_model.predict(X.head())
```

```python
df['Price'].head()
```

```python
# using train test split method

from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 0)
```

```python
house_model_new = DecisionTreeRegressor()
```

```python
house_model_new.fit(train_X, train_y)
```

```python
# testing the model
house_model_new.predict(train_X.head())
```

```python
train_y.head()
```

```python
# predicting the test data
predicted_on_test = house_model_new.predict(test_X.head())
predicted_on_test
```

```python
test_y.head()
```

```python
# evaluating the model performance
# Mean Absolute Error (MAE)

from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test_y, house_model_new.predict(test_X))
```

```python
# Mean Squared Error (MSE)

from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(test_y, house_model_new.predict(test_X))
```

```python
# Root Mean Squared Error (RMSE)

from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(test_y, house_model_new.predict(test_X))
```

```python
# R2 Score
from sklearn.metrics import r2_score
```

```python
r2_score(test_y, house_model_new.predict(test_X))
```

```python
# Trying the new algorithm
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
```

```python
forest_model = RandomForestRegressor(random_state=1)
```

```python
# fitting the data into the model

forest_model.fit(train_X, train_y)
```

```python
hp3 = forest_model.predict(test_X)
```

```python
hp3
```

```python
test_y
```

```python
mean_absolute_error(test_y, hp3)
```

```python
root_mean_squared_error(test_y, hp3)
```
