# 🤖 Unsupervised Machine Learning

## 📌 What is Unsupervised Learning?
Unsupervised Machine Learning is a type of ML where the model **learns patterns from unlabeled data** without predefined outputs.

## 🔹 Key Features:
- No labeled data required.
- Identifies hidden structures and relationships in data.
- Used for clustering, association, and dimensionality reduction.

## 📊 Common Types:
### 1️⃣ **Clustering** 🏷️
   - Groups similar data points together.
   - Example: **K-Means, DBSCAN, Hierarchical Clustering**  
   - **Use Case:** Customer segmentation in marketing.

### 2️⃣ **Dimensionality Reduction** 📉
   - Reduces data size while preserving important features.
   - Example: **PCA (Principal Component Analysis), t-SNE**  
   - **Use Case:** Image compression, data visualization.

### 3️⃣ **Association Rule Learning** 🔗
   - Discovers relationships between variables.
   - Example: **Apriori, FP-Growth**  
   - **Use Case:** Market basket analysis (e.g., "People who buy bread also buy butter").

## 🛠️ Example Applications:
- **Anomaly Detection** (e.g., Fraud detection)
- **Image Segmentation** (e.g., Object detection in images)
- **Document Clustering** (e.g., Topic modeling for news articles)

## ⚡ Advantages:
✔️ Can handle large amounts of data.  
✔️ Finds hidden patterns without human intervention.  
✔️ Helps in exploratory data analysis.

## ❌ Challenges:
❌ Hard to evaluate accuracy (no labeled data).  
❌ Can be sensitive to hyperparameters.  
❌ May not always provide meaningful insights.

---

💡 **Conclusion:**  
Unsupervised ML is **powerful for discovering hidden insights** from data, making it essential for pattern recognition, clustering, and exploratory analysis. 🚀


# 🔍 Difference Between Supervised and Unsupervised Machine Learning

| Feature               | Supervised Learning 🏫 | Unsupervised Learning 🤖 |
|----------------------|--------------------|----------------------|
| **Definition**        | Learns from labeled data (input-output pairs) | Learns from unlabeled data, finding patterns |
| **Output**           | Predicts known outcomes (classification/regression) | Finds hidden structures (clustering, association) |
| **Example Algorithms** | Linear Regression, KNN, Decision Trees, SVM | K-Means, DBSCAN, PCA, Autoencoders |
| **Example Use Cases** | Spam detection, Disease prediction, Stock price forecasting | Customer segmentation, Anomaly detection, Market basket analysis |
| **Dependency on Labels** | Requires labeled data for training | No labeled data needed |
| **Complexity**       | Easier to interpret | Harder to interpret (no predefined labels) |
| **Real-world Examples** | Email spam detection (Spam/Not Spam) | Grouping customers based on shopping behavior |

# 🔍 Clustering and Dimensionality Reduction in Machine Learning

## 📌 1. What is Clustering?  
Clustering is an **unsupervised learning** technique that groups similar data points together **without labeled outputs**.

### 🔹 Key Features:
- Finds **hidden patterns** in data.
- Groups data points based on **similarity**.
- Used for **customer segmentation, anomaly detection**, etc.

### 🛠️ Common Clustering Algorithms:
1. **K-Means Clustering** – Divides data into **K clusters** using centroid-based grouping.  
2. **Hierarchical Clustering** – Builds a **tree of clusters** (dendrogram).  
3. **DBSCAN (Density-Based Clustering)** – Groups **dense areas** and detects anomalies.  

### 📊 Example Use Cases:
✔ **Customer Segmentation** – Grouping shoppers based on buying patterns.  
✔ **Image Segmentation** – Identifying objects in images.  
✔ **Anomaly Detection** – Detecting fraud in transactions.  

---

## 📌 2. What is Dimensionality Reduction?  
Dimensionality reduction is a technique used to **reduce the number of input features** while retaining essential information.

### 🔹 Why Use Dimensionality Reduction?
- **Reduces computational cost** for large datasets.  
- **Removes noise and redundancy** from data.  
- **Improves model performance** by avoiding the "curse of dimensionality."  

### 🛠️ Common Dimensionality Reduction Techniques:
1. **Principal Component Analysis (PCA)** – Converts high-dimensional data into a lower-dimensional space while preserving variance.  
2. **t-SNE (t-Distributed Stochastic Neighbor Embedding)** – Used for **visualizing high-dimensional data** in 2D/3D.  
3. **LDA (Linear Discriminant Analysis)** – Used for **supervised dimensionality reduction** to maximize class separability.  

### 📊 Example Use Cases:
✔ **Data Visualization** – Reducing 100D data to 2D for plotting.  
✔ **Image Compression** – Removing redundant pixel information.  
✔ **Speeding Up Machine Learning Models** – Faster training time with fewer features.  

---

## 🔥 Key Differences: Clustering vs. Dimensionality Reduction

| Feature              | Clustering 🏷️ | Dimensionality Reduction 📉 |
|---------------------|-------------|--------------------------|
| **Goal**            | Group similar data points | Reduce feature count while retaining info |
| **Type of ML**      | Unsupervised Learning | Can be used in both Supervised & Unsupervised ML |
| **Examples**        | K-Means, DBSCAN, Hierarchical Clustering | PCA, t-SNE, LDA |
| **Use Case**        | Customer segmentation, fraud detection | Image compression, feature selection |

---

💡 **Conclusion:**  
- **Clustering** is used for **grouping similar data** without labels.  
- **Dimensionality Reduction** is used for **simplifying data** while keeping essential information.  
- Both help in **handling large datasets efficiently** in machine learning. 🚀

```python
ms= MinMaxScaler()
X= ms.fit_transform(X)
X= pd.DataFrame(X, columns=features)
```

```python
#initialize k means clustering 
from sklearn.cluster import KMeans
```

```python
# create a kmeans model 
kmeans= KMeans(n_clusters= 3, random_state= 42)
```

```python
kmeans
```

```python
kmeans.fit(X)
```

```python
kmeans.inertia_
```

```python
# evaluate clustering using silhouette_score 
from sklearn.metrics import silhouette_score
```

```python
silhouette_avg= silhouette_score(X, kmeans.labels_)
```

```python
silhouette_avg
```

```python
#silhouette score visualization
# Number of clusters
from sklearn.metrics import silhouette_samples, silhouette_score
silhouette_vals = silhouette_samples(X, kmeans.labels_)

n_clusters = kmeans.n_clusters

# Create a figure
plt.figure(figsize=(10, 6))

y_lower = 10
for i in range(n_clusters):
    # Aggregate the silhouette scores for samples in cluster i
    ith_cluster_silhouette_vals = silhouette_vals[kmeans.labels_ == i]
    ith_cluster_silhouette_vals.sort()

    # Compute the size of the cluster
    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    # Fill the silhouette plot for the cluster
    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_vals,
        alpha=0.7,
        label=f'Cluster {i + 1}'
    )

    # Label the silhouette scores
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i + 1))
    y_lower = y_upper + 10  # 10 for the 0 samples gap between clusters

# Vertical line for the average silhouette score
plt.axvline(x=silhouette_avg, color="red", linestyle="--", label="Average Silhouette")

# Plot labels
plt.title("Silhouette Plot for K-Means Clustering", fontsize=15)
plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster Label")
plt.legend()
plt.show()
```
